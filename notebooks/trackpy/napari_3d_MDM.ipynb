## Nuclei detection
We use a simple gaussian filtering followed by local maxima detection to find all the nuclei and create an array with time, x, y,z positions.

```python
blue = blue.compute()
all_dots = []
# filtered = np.zeros(blue.shape)
for t in range(blue.shape[0]):
    imG = skimage.filters.gaussian(blue[t,:,:],sigma =2,
                               preserve_range=True)
#     filtered[t,:,:] = imG
    locmax = skimage.feature.peak_local_max(imG,min_distance=45,
                                            threshold_abs=5000)
    newloc = np.stack((t*np.ones(locmax.shape[0]),
                       locmax[:,0],locmax[:,1]),axis =1)
    all_dots.append(newloc)
spots = np.concatenate(all_dots) 

with open('nuclei.npy', 'wb') as f:
    np.save(f, spots)

```

```python
%matplotlib qt5


def find_nuclei(I,gaussianblur=2,remove_small=50):
    import numpy as np
    from skimage import morphology
    from skimage.filters import gaussian, threshold_local, threshold_mean
    from skimage.feature import peak_local_max
    from skimage.morphology import remove_small_objects
    import skimage.segmentation
    from skimage.segmentation import watershed
    from scipy import ndimage
    import glob
    import scipy.ndimage as ndi

    # Functional parameters
    # IntensityBlurRad = float(args.blurad) # 0.0
    # RadThr = float(args.radthr) # 75
    # Thr = float(args.intthr)   # -10
    # DistanceBlurRad = float(args.spltnt) # 3
    # MinSize = int(args.minsize) # 25
#     IntensityBlurRad = 1# 0.0
#     RadThr = 15
#     Thr =  0# -10
#     DistanceBlurRad = 3
#     MinSize = 2


#     I = blue[0].compute()

       # Gaussian filter

    if IntensityBlurRad >= 1:
        I = gaussian(I, sigma=IntensityBlurRad,  preserve_range=True)

    mask = I >= threshold_mean(I)*2

    sizestruct = (9,9)
    cell = ndi.binary_opening(mask,
            structure=np.ones(sizestruct))
    mask = ndi.binary_closing(mask,iterations=2)
    mask = ndi.binary_fill_holes(mask,structure=np.ones((5,5)))
    mask = ndi.binary_erosion(mask,structure=np.ones((3,3)))


    mask = morphology.remove_small_objects(mask,remove_small)

    distance = ndimage.distance_transform_edt(mask)
    sizestruct = (25,25)
    local_maxi = peak_local_max(distance, indices=False, footprint=np.ones(sizestruct),
    #     threshold_abs=np.mean(I)*3,
        min_distance=25,
                                labels=mask)
    markers = ndi.label(local_maxi)[0]
    labels = watershed(-distance, markers, mask=mask)

    coords = ndi.measurements.center_of_mass(blue[0].compute(),labels,np.arange(1,labels.max()+1))
    
    return coords#,labels
```

```python
all_dots = []

for i in range(blue.shape[0]):
    print(i)
    all_dots.append(find_nuclei(blue[i].compute(),gaussianblur=2,remove_small=50))
    
newdots = []
for t in range(blue.shape[0]):
    locmax = np.stack(all_dots[t])
#     newdots.append(np.stack(locmax))
    newdots.append(np.stack((t*np.ones(locmax.shape[0]),
                       locmax[:,0],locmax[:,1]),axis =1))
spots = np.concatenate(newdots)
# spots = np.asarray(all_dots)
```

## Tracking

Finally we use [trackpy](https://soft-matter.github.io/trackpy/v0.3.2/) to track the different nuclei. For that we just have to turn our time,x,y,z array into a Pandas dataframe and then call trackpy:

```python
spots_pd = pd.DataFrame(spots,columns=['frame','y','x'])
traj = trackpy.link_df(spots_pd,search_range=12)
with trackpy.PandasHDFStore('TrajectoriesNuclei.h5') as s:
    s.put(traj)
```

We just exclude very short tracks to make the plot more readable:

```python
all_traj = []
for ind in traj.particle.unique():
    curr_traj = traj[traj.particle==ind][['y','x']].values
    if curr_traj.shape[0]>5:
        all_traj.append(curr_traj)
```
