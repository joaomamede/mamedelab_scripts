```python
import napari
import numpy as np
import pandas as pd
import trackpy as tp

from glob import glob
from skimage.io import imread
from skimage.io.collection import alphanumeric_key
```

```python
SAMPLE_DATA_PATH = "/home/quantumjot/Downloads/trackpy-examples-master/sample_data/bulk_water/*.png"
```

```python
filenames = sorted( glob(SAMPLE_DATA_PATH), key=alphanumeric_key )
```

```python
def gray(fn):
    img = imread(fn)
    return img[:, :, 1]

stack = np.stack([gray(fn) for fn in filenames], axis=0)
```

```python
f = tp.batch(stack[:300], 11, minmass=20, invert=True)
```

```python
t = tp.link(f, 5, memory=3)
```

```python
t1 = tp.filter_stubs(t, 25)
```

```python
t2 = t1[((t1['mass'] > 50) & (t1['size'] < 2.6) &
         (t1['ecc'] < 0.3))]
```

```python
# lexsort data
tm = t2.reset_index(drop=True)
tm = tm.sort_values(['particle', 'frame'], ignore_index=True)
```

```python
tm.head()
```

```python
data = tm.loc[:, ['particle','frame','y','x']]
data = data.to_numpy()
```

```python
properties = tm.loc[:, ['mass', 'size', 'ecc', 'signal', 'raw_mass', 'ep']]
```

```python
# with napari.gui_qt():
%gui qt
viewer = napari.Viewer()
viewer.add_image(stack)
viewer.add_points(data[:, 1:], properties=properties, size=1, name='particles', face_color='white')
viewer.add_tracks(data, properties=properties, name='tracks', blending='translucent')
```
